```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
genes = pd.read_csv("data_set_ALL_AML_train.csv")
```

```python
genes.columns
```

```python
expression_values = [column for column in genes if "call" not in column]
```

creating list of columns' names only with expression values. where 'coll' word is not found
then this list of columns names is given to a DataFrame genes obj as a key, so it gets only the columns (the Series obj with expression values)

```python
genes = genes[expression_values]
```

```python
genes[:10]
```

```python
genes1 = genes.drop(["Gene Description", "Gene Accession Number"], axis=1)
```

```python
genes1[:10]
```

```python
genes1 = genes1.T
```

```python
genes1.head()
```

```python
from sklearn.decomposition import FactorAnalysis
from sklearn import preprocessing
```

```python
genes_normal = preprocessing.scale(genes1)
type(genes1)
```

```python
factor = FactorAnalysis(n_components=5).fit(genes_normal)
```

```python
factor_arr = pd.DataFrame(factor.components_, columns=genes1.columns)
```

```python
factor_arr
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=20).fit(genes_normal)
```

```python
pca
```

```python
pca.explained_variance_ratio_
```
